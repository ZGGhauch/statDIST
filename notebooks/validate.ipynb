```python
import numpy as np
import matplotlib.pyplot as plt
from seaborn import histplot 
np.random.seed(999)

import sys
sys.path.append('../statDIST/')
from metrics import Metrics
from metrics import StatDist
from divergences import Divergence
```

```python
def sample_distribution(N,D,mean,cov):
    rvX= np.random.multivariate_normal(mean=np.zeros(D)+mean,
                                      cov = cov,
                                      size = N,
                                      check_valid='warn',
                                      tol = 1e-8)
    return (rvX)
```

```python
X = np.random.normal(0.0,1.0,10000)
Divergence(X, X, True, 200).plot_density_kde(X, 300)
```

## Bhattacharyya distance

```python
x_grid = np.linspace(0.0, 50., 1000)
fig, ax = plt.subplots(4, 4, sharey=True, figsize=(14,14))
fig.subplots_adjust(wspace=0.0)
fig.suptitle('Bhattacharyya Distance', fontsize=16)

for i in range(4):
	for j in range(4):
		#np.random.seed(i*j)
				
		n1,n2 = 1000,1000
		d1,d2 = 1,1
		mean1 = np.random.uniform(5.,45.)
		mean2 = np.random.uniform(5.,45.)
		xi1 = np.random.normal(1.,0.1,d1)
		xi2 = np.random.normal(1.,0.2,d2)
		
		X = sample_distribution(N=n1, D=d1, mean=mean1, cov=np.eye(d1)*xi1)
		#denX = StatDist.compute_density_kde(X.flatten())
		Y = sample_distribution(N=n2, D=d2, mean=mean2, cov=np.eye(d2)*xi2)
		#denY = StatDist.compute_density_kde(Y.flatten())	

		# plot kernel density estimates
		z = histplot(X, label='X', color='orange', kde=True, ax=ax[i,j], legend=False, stat='density', alpha = 1.0)
		z = histplot(Y, label='Y', color='skyblue', kde=True, ax=ax[i,j], legend=False, stat='density', alpha = 0.1)
		#ax[i,j].set_xlim(0.0, 50.0)

		# Bhattacharyya distance
		bhatD = np.round(Divergence(X, Y, True, 200).compute_bhat_distance(),2)
		z.set_title(str(bhatD))
```
